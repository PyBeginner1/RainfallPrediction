```python
data['RainToday'] = data['RainToday'].fillna('No')
```

```python
data['RainTomorrow'] = data['RainTomorrow'].fillna('No')
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()

label_encoder_columns = ['RainToday', 'RainTomorrow']
for column in label_encoder_columns:
  data[column] = encoder.fit_transform(data[column])
```

```python
data['RainToday'].unique()
```

```python
data['WindGustDir'].unique()
```

```python
data['WindGustDir'].value_counts()
```

```python
def add_columns_prefixes(data, column, prefix):
  return data[column].apply(lambda x: prefix + str(x))
```

```python
add_columns_prefixes(data, 'Location', 'loc_')
```

```python
data['WindDir3pm'] = add_columns_prefixes(data, 'WindDir3pm', '3_')
data['WindDir9am'] = add_columns_prefixes(data, 'WindDir9am', '9_')
```

```python
data
```

```python
def onehot_encoder(data, columns):
  for column in columns:
    dummy = pd.get_dummies(data[column])
    data = pd.concat([data, dummy], axis = 1)
    data.drop(column, axis = 1, inplace = True)
  return data
```

```python
categorical_features = ['Location', 'WindGustDir', 'WindDir9am', 'WindDir3pm']
data = onehot_encoder(data, categorical_features)
```

```python
data
```

```python
data.info()
```

```python
data.isna().sum()
```

Filling empty numerical Values with mean

```python
def impute_means(data, columns):
  for column in columns:
    data[column] = data[column].fillna(data[column].mean())
```

```python
numerical_features=['MinTemp','MaxTemp', 'Rainfall', 'Evaporation', 'Sunshine', 'WindGustSpeed', 'WindSpeed9am', 'WindSpeed3pm', 'Humidity9am', 'Humidity3pm', 'Pressure9am', 'Pressure3pm','Cloud9am',
                    'Cloud3pm', 'Temp9am', 'Temp3pm' ]

impute_means(data, numerical_features)
```

```python
data.isna().sum().sum()
```

```python
X = data.drop('RainTomorrow', axis = 1)
Y = data['RainTomorrow']
```

```python
X
```

```python
from sklearn.preprocessing import RobustScaler
```

```python
scaler = RobustScaler()
X = pd.DataFrame(scaler.fit_transform(X), columns = X.columns)
```

```python
X
```

Train - test split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model

```python
import tensorflow as tf 
```

```python
inputs = tf.keras.Input(shape=(116,))
x = tf.keras.layers.Dense(16, activation = 'relu')(inputs)
x = tf.keras.layers.Dense(16, activation = 'relu')(x)
outputs = tf.keras.layers.Dense(2, activation = 'softmax')(x)

model = tf.keras.Model(inputs= inputs, outputs = outputs)
```

```python
model.summary()
```

```python
model.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)
```

```python
EPOCHS = 9
```

```python
history = model.fit(
    X_train,
    Y_train,
    epochs = EPOCHS,
    batch_size = 32,
    verbose = 1,
    validation_split = 0.2
)
```

```python
plt.figure(figsize=(10,10))

plt.plot(range(EPOCHS), history.history['loss'], color = 'b')
plt.plot(range(EPOCHS), history.history['val_loss'], color = 'r')
```

Finding epoch with low val loss

```python
np.argmin(history.history['val_loss'])
```
